### Tabel

Ketika Anda menyusun data ke dalam format baris dan kolom, maka Anda telah berhasil membuat sebuah tabel. Tabel merupakan salah satu media visualisasi data yang sederhana dan sering kita temukan.

Data dalam tabel biasanya dikategorikan dalam baris atau kolom tertentu yang kemudian dapat diurutkan dengan mudah, misal diurutkan secara menaik (ascending) atau menurun (descending). Hal tersebut memudahkan kita untuk menyusun data sesuai dengan perintah yang diinginkan.

Bayangkan Anda memiliki data buah-buahan berikut:

| Nama Buah  | Jumlah |
|------------|--------|
| Apel       | 50     |
| Jambu      | 35     |
| Melon      | 40     |
| Semangka   | 10     |
| Mangga     | 25     |
| Jeruk      | 100    |
| **TOTAL**  | **260**|

Dari tabel di atas, Anda dapat membaca data buah-buahan dengan lebih efektif dan mengetahui jumlah total buah. Kita dapat mengurutkan data tersebut secara ascending berdasarkan kolom Nama Buah sehingga akan tampak seperti berikut.

| Nama Buah  | Jumlah |
|------------|--------|
| Apel       | 50     |
| Jambu      | 35     |
| Jeruk      | 100    |
| Mangga     | 25     |
| Melon      | 40     |
| Semangka   | 10     |
| **TOTAL**  | **260**|

Selain berdasarkan Nama Buah, kita juga bisa mengurutkan data dari jumlah buah yang paling banyak atau paling sedikit.

**Aturan dasar penulisan tabel:**
- **Penulisan Judul:** Pastikan judul sudah mencakup isi tabel, gunakan font yang jelas dan mudah dibaca.
- **Simpel:** Desain tabel sederhana agar pembaca fokus pada data.
- **Penjelasan Simbol:** Jelaskan simbol atau istilah pada catatan kaki tabel.
- **Penekanan:** Fokuskan perhatian pembaca pada pokok data, misal dengan warna berbeda atau menambahkan total.
- **Sumber Tabel:** Jika tabel bukan milik Anda, sertakan sumber di catatan kaki.

### Diagram

Diagram merupakan representasi data dalam bentuk grafik. Berikut beberapa tipe diagram yang sering digunakan:

#### Diagram Batang

Diagram batang menunjukkan perbandingan angka pada kategori tertentu. Sumbu X menunjukkan kategori data, sumbu Y menunjukkan skala nilai. Pastikan jumlah batang tidak terlalu banyak agar label terbaca jelas. Gunakan warna yang konsisten dan label yang mudah dibaca.

#### Diagram Garis

Diagram garis menyajikan perubahan data dalam periode waktu tertentu. Sumbu X untuk interval waktu, sumbu Y untuk kuantitas. Titik-titik data dihubungkan dengan garis untuk melihat tren.

*Contoh data penjualan sepeda tahun 2019:*

| Bulan      | Jumlah |
|------------|--------|
| Januari    | 150    |
| Februari   | 130    |
| Maret      | 180    |
| April      | 190    |
| Mei        | 250    |
| Juni       | 200    |
| Juli       | 230    |
| Agustus    | 250    |
| September  | 280    |
| Oktober    | 290    |
| November   | 300    |
| Desember   | 350    |
| **Total**  | **2800**|

#### Diagram Lingkaran

Diagram lingkaran menggambarkan persentase data dalam bentuk irisan seperti pizza. Total irisan adalah 100% atau 360 derajat.

*Contoh survei hobi siswa:*

| Hobi         | Jumlah |
|--------------|--------|
| Sepak Bola   | 17     |
| Basket       | 3      |
| Bulu Tangkis | 5      |
| Lainnya      | 10     |
| **Total**    | **35** |

*Contoh diagram lingkaran:*

![Contoh Diagram Lingkaran](202004121700325f986909b2ef7d5b175bb32227995a23.jpeg)

**Rumus diagram lingkaran:**
- Dalam persen:  
    `(Jumlah data kategori tertentu / Total data keseluruhan) x 100`
- Dalam derajat:  
    `(Jumlah data kategori tertentu / Total data keseluruhan) x 360`

**Aturan dasar diagram lingkaran:**
- Sertakan judul.
- Pastikan total data 100% (persen) atau 360° (derajat).
- Label kategori dan nilai harus jelas.
- Gunakan warna berbeda untuk tiap irisan.
- Maksimal 5 irisan agar mudah dibaca.

### Ringkasan Visualisasi Data dalam Bisnis

Visualisasi data sangat penting dalam bisnis untuk menyampaikan informasi yang kompleks secara efektif dan menghindari kesalahan analisis. Berikut tipe visualisasi data yang umum digunakan di bidang bisnis:

- **Scorecard**  
    Menampilkan satu jenis data spesifik, seperti pendapatan atau KPI, untuk memantau kemajuan terhadap target.

- **Dashboard**  
    Gabungan berbagai visualisasi yang merangkum informasi bisnis. Dashboard biasanya terdiri dari beberapa scorecard dan visualisasi lain yang saling berhubungan.

- **Report**  
    Laporan yang menggunakan visualisasi data untuk merangkum dan menyampaikan informasi penting perusahaan dalam periode tertentu.

- **Analytic Report**  
    Laporan analitik yang fokus pada analisis data kualitatif dan kuantitatif untuk mendukung pengambilan keputusan bisnis, mulai dari identifikasi masalah hingga solusi.

Visualisasi ini membantu perusahaan memahami data secara cepat dan akurat untuk pengambilan keputusan yang lebih baik.

## Tools dalam Visualisasi Data

Dalam membuat visualisasi data, terdapat beberapa tools gratis yang dapat digunakan untuk mempermudah proses pembuatan grafik dan diagram:

- **Tableau Public**  
    Layanan gratis yang memungkinkan pengguna mempublikasikan visualisasi data ke web. Visualisasi dapat dibagikan ke media sosial, blog, atau diunduh oleh pengguna lain. Pembuatan visualisasi dilakukan melalui aplikasi Tableau Desktop Public Edition tanpa perlu keahlian pemrograman.

- **Google Sheets**  
    Aplikasi spreadsheet online dari Google yang menyediakan fitur standar seperti Microsoft Excel. Google Sheets dapat digunakan untuk membuat visualisasi sederhana seperti diagram batang, garis, dan lingkaran. Layanan ini gratis, namun memerlukan akun Google.

- **Microsoft Excel**  
    Aplikasi spreadsheet berbayar dari Microsoft yang sangat populer. Selain untuk manajemen data dan perhitungan numerik, Excel juga dapat digunakan untuk membuat berbagai visualisasi data sederhana seperti grafik garis, batang, dan lingkaran. Instalasi aplikasi diperlukan sebelum digunakan.

Dengan memanfaatkan tools di atas, proses visualisasi data menjadi lebih mudah dan efisien, baik untuk pemula maupun profesional.

### Ringkasan Konteks Data

- **Data mencerminkan kehidupan nyata yang kompleks.** Memahami konteks data sangat penting agar data dapat dipahami dan disampaikan dengan tepat.
- **Kenali audiens.** Mengetahui siapa yang akan menerima data membantu menentukan cara penyampaian yang efektif dan relevan.
- **Tentukan poin penting.** Pilih data yang relevan sesuai kebutuhan audiens agar informasi yang disampaikan tidak berlebihan atau kurang tepat sasaran.
- **Sesuaikan gaya penyampaian.** Gunakan bahasa dan metode yang sesuai dengan tingkat pemahaman audiens, baik umum maupun profesional.
- **Gunakan narasi jika perlu.** Menyampaikan data dengan cerita dapat membantu audiens memahami makna di balik angka, bukan sekadar membaca data mentah.
- **Inti pertanyaan sebelum menyajikan data:**
    1. Siapa audiensnya?
    2. Data apa yang ingin disampaikan?
    3. Bagaimana cara menyampaikannya?
