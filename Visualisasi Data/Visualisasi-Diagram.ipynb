### Ringkasan Visualisasi Data menggunakan Diagram

- **Diagram** adalah representasi visual dari data untuk memudahkan pemahaman informasi.
- Diagram membantu mempercepat proses memahami data dibandingkan hanya melihat tabel/baris-kolom.
- Grafik mempersingkat waktu pemahaman dan mendukung hipotesis secara visual.

#### Jenis-jenis Diagram:
1. **Bar/Column Chart**  
    Menampilkan data dalam bentuk persegi panjang (batang) yang panjangnya sesuai nilai data.

2. **Line Chart**  
    Menampilkan data sebagai titik-titik yang dihubungkan garis lurus, cocok untuk melihat tren.

3. **Area Chart**  
    Mirip Line Chart, tetapi area di bawah garis diwarnai untuk menonjolkan volume data.

4. **Pie Chart**  
    Menampilkan data dalam bentuk lingkaran, tiap irisan menunjukkan proporsi bagian dari keseluruhan.

5. **Scatter Chart**  
    Menggunakan titik-titik untuk menunjukkan nilai pada dua sumbu, baik untuk melihat hubungan antar variabel.

6. **Map Chart**  
    Memvisualisasikan data pada peta geografis, warna menunjukkan intensitas atau kepadatan data.

Diagram memudahkan penyerapan informasi dan tersedia dalam berbagai bentuk di aplikasi seperti Google Sheets.

### Ringkasan Kesalahan Umum dalam Visualisasi Data

Berikut adalah beberapa kesalahan umum yang sering terjadi dalam visualisasi data beserta penjelasannya:

| Kesalahan Umum                        | Penjelasan                                                                                                   | Contoh/Ilustrasi                                                                                   |
|----------------------------------------|-------------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------------|
| **Nilai Persentase Tidak Sesuai**      | Total persentase tidak 100% atau porsi irisan tidak sesuai nilai.                                           | Pie chart dengan total <100% atau irisan 42% lebih besar dari 56%.                                 |
| **Terlalu Banyak Data**                | Memasukkan terlalu banyak kategori membuat visualisasi sulit dibaca dan membingungkan.                      | Pie chart dengan banyak irisan, sulit membedakan nilai masing-masing.                              |
| **Tidak Mengikuti Standar Penulisan**  | Sumbu grafik tidak sesuai standar, misal sumbu Y terbalik (0 di atas, nilai besar di bawah).                | Grafik tren dengan sumbu Y terbalik, menyebabkan salah persepsi.                                   |
| **Terdapat Sumbu yang Terpotong**      | Sumbu tidak dimulai dari 0 sehingga perbedaan data tampak lebih besar dari kenyataan.                       | Bar chart dengan sumbu Y mulai dari 34%, bukan 0.                                                  |
| **Penggunaan Grafik 3D yang Tidak Sesuai** | Grafik 3D dapat menimbulkan bias visual dan menyulitkan pembaca memahami data.                              | Pie chart/Bar chart 3D yang membuat proporsi data tampak berbeda dari data aslinya.                |
| **Sukar Dibandingkan**                 | Visualisasi tidak memudahkan pembaca membandingkan data, baik antar kategori maupun antar grafik.            | Banyak pie chart kecil yang sulit dibandingkan satu sama lain.                                     |

#### Tips Menghindari Kesalahan:
- Pastikan total persentase pada pie chart adalah 100%.
- Gunakan maksimal 4 kategori pada pie chart, selebihnya gunakan kategori "Lainnya" dengan keterangan jelas.
- Hindari memasukkan terlalu banyak data dalam satu grafik.
- Ikuti standar penulisan sumbu (sumbu Y dari bawah ke atas, mulai dari 0).
- Hindari penggunaan grafik 3D jika tidak diperlukan.
- Pilih jenis grafik yang memudahkan perbandingan data.

### Ringkasan Dokumentasi Data (Data Provenance)

- **Dokumentasi data** (data provenance) adalah proses mencatat asal-usul, sumber, dan perubahan data.
- Penting untuk memastikan keabsahan, validitas, dan kepercayaan terhadap data, serta memudahkan penggunaan ulang di masa depan.
- Metadata adalah bagian dari dokumentasi data yang mencakup informasi tentang sumber, metodologi, pemilik, dan kualitas data.
- Contoh dokumentasi data:
    - Metadata pada dataset di Kaggle yang mencantumkan sumber dan proses pembuatan data.
    - Version history di Google Sheets yang merekam perubahan data, tanggal, dan nama pengubah, serta memungkinkan pemulihan versi sebelumnya.
- Dokumentasi data membantu menilai keaslian, kredibilitas, dan reprodusibilitas data.

### Pentingnya Dokumentasi Data

Dokumentasi data sangat penting untuk memastikan **kepercayaan**, **kredibilitas**, dan **reproduksibilitas** data. Dokumentasi sumber data yang baik memungkinkan pengguna memahami asal-usul, proses, dan perubahan data, meskipun mereka bukan pembuat data.

#### Peran Pembuat Data
- Mengonfigurasikan instrumen/simulasi untuk mengumpulkan data primer.
- Menerapkan metodologi/proses untuk mengekstraksi, mengubah, dan menganalisis data.

---

### Pengelolaan Sumber Data & Metadata

- **Provenance** dicatat dalam metadata, seperti tanggal pembuatan, pemilik, perangkat lunak, metode pemrosesan, dsb.
- Pengelolaan data yang baik menjadi dasar dokumentasi data yang akurat.
- Contoh: Blockchain pada supply chain untuk melacak asal-usul ikan melalui barcode.

**Metadata**: Lokasi penangkapan, nelayan, waktu, dsb.  
**Data**: Ikan itu sendiri.

---

### Ringkasan: Bercerita dengan Data

Bercerita dengan data bertujuan menyampaikan hasil visualisasi yang mendukung hipotesis secara efektif agar dapat memberikan insight dan membantu pengambilan keputusan.

---

#### Perbedaan Exploratory vs Explanatory

| Jenis Analisis         | Tujuan Utama                                      | Cara Penyampaian                |
|------------------------|---------------------------------------------------|----------------------------------|
| Exploratory Analysis   | Menjelajahi data bersama audiens                  | Membimbing audiens menjelajah    |
| Explanatory Analysis   | Menyampaikan poin utama & implikasi data langsung | Fokus pada inti cerita/data      |

Mayoritas situasi lebih cocok menggunakan explanatory analysis.

---

#### Tiga Pertanyaan Penting Explanatory Analysis

| Pertanyaan                                    | Penjelasan Singkat                                                                 |
|-----------------------------------------------|------------------------------------------------------------------------------------|
| Kepada siapa kita berkomunikasi?              | Kenali audiens, kebutuhan informasi, dan hubungan dengan audiens                   |
| Bagaimana kita berkomunikasi dengan audiens?  | Pilih alat komunikasi, sesuaikan detail, dan tentukan nada penyampaian             |
| Bagaimana data menegaskan maksud kita?        | Gunakan data untuk mendukung cerita, terapkan metode 5W (What, Who, When, Where, Why) |

---

| Unsur 5W  | Contoh Pernyataan                                                                                   |

---

Gunakan pendekatan ini agar cerita data Anda jelas, terstruktur, dan meyakinkan audiens.


### Visualisasi yang Efektif

Terdapat banyak jenis visualisasi data, namun hanya sebagian yang benar-benar efektif menjawab kebutuhan komunikasi data. Berikut beberapa visualisasi yang sering digunakan dan tips penggunaannya:

---

#### **1. Visualisasi Berbentuk Teks**
Gunakan teks sederhana untuk menampilkan satu atau dua angka penting. Tampilkan angka secara menonjol dan tambahkan kalimat pendukung agar pesan lebih jelas dan mengurangi potensi salah tafsir.

---

#### **2. Tabel**
Tabel cocok untuk data dengan banyak angka atau unit pengukuran berbeda. Tabel memudahkan pembaca membandingkan nilai antar baris/kolom sesuai kebutuhan. Hindari garis tepi tebal atau bayangan berlebihan; gunakan garis tipis atau ruang putih agar fokus tetap pada data.

---

#### **3. Heatmap**
Heatmap menggabungkan detail angka pada tabel dengan visualisasi warna. Pewarnaan sel membantu pembaca lebih cepat mengidentifikasi nilai tertinggi/terendah tanpa harus membaca satu per satu.

---

#### **4. Grafik**
Grafik mempercepat pemahaman informasi dibandingkan tabel. Empat kategori utama grafik:

- **Titik (Scatterplot):** Menunjukkan hubungan antara dua variabel (misal: jarak tempuh vs biaya/km). Cocok untuk analisis korelasi.
- **Garis (Line Chart & Slopegraph):** Menampilkan perubahan data kontinu dari waktu ke waktu. Slopegraph fokus pada perubahan antara dua titik.
- **Batang (Bar Chart):** Paling umum dan mudah dibaca. Cocok untuk membandingkan antar kategori. Pastikan sumbu Y mulai dari 0 dan lebar batang lebih besar dari jarak antar batang.
    - **Vertikal/Horizontal:** Pilih horizontal jika nama kategori panjang.
    - **Bertumpuk (Stacked):** Bandingkan total dan sub-komponen, namun bisa membingungkan jika terlalu banyak kategori.
    - **Waterfall:** Menunjukkan perubahan bertahap dari satu nilai ke nilai lain.
- **Area:** Gunakan jika perbedaan nilai sangat besar. Visualisasi area efektif jika ingin menampilkan proporsi besar vs kecil secara ringkas.

---

**Tips Umum:**
- Pilih visualisasi sesuai tujuan dan jumlah data.
- Fokus pada kejelasan pesan, bukan dekorasi.
- Gunakan warna dan garis bantu secukupnya untuk memperjelas, bukan mengalihkan perhatian.

Visualisasi yang efektif akan membantu audiens memahami data dengan cepat dan tepat.

### Ringkasan Prinsip Gestalt dalam Persepsi Visual

Prinsip Gestalt membantu membuat visualisasi data lebih mudah dipahami dengan mengelompokkan elemen berdasarkan pola dan hubungan tertentu. Berikut ringkasan enam prinsip utama beserta penjelasan dan contoh penerapannya:

| Prinsip      | Penjelasan Singkat                                                                 | Contoh Visualisasi/Penerapan                                                                                   |
|--------------|------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------------------------|
| **Proximity** (Kedekatan)   | Objek yang berdekatan dianggap satu kelompok.                                   | Titik-titik yang berdekatan terlihat sebagai satu grup.                                                        |
| **Similarity** (Kesamaan)   | Objek dengan warna/bentuk/ukuran sama dianggap satu kelompok.                   | Warna/bentuk serupa pada grafik menunjukkan kategori yang sama, meski tidak berdekatan.                        |
| **Enclosure** (Pembeda)     | Objek dalam satu area/border dianggap satu kelompok.                            | Area berbayang pada grafik membedakan data aktual dan prediksi.                                                |
| **Closure** (Penutupan)     | Otak melengkapi bentuk yang tidak utuh menjadi satu kesatuan.                   | Grafik tanpa border tetap dianggap satu kesatuan; desain sederhana menonjolkan data.                           |
| **Continuity** (Kesinambungan) | Otak membentuk pola/garis berkesinambungan meski tidak eksplisit.                | Sumbu-y dihapus, tetapi batang grafik tetap terlihat berbaris rapi.                                            |
| **Connection** (Koneksi)    | Objek yang terhubung secara fisik dianggap satu kelompok, lebih kuat dari proximity/similarity. | Garis penghubung pada grafik garis membuat titik-titik terlihat sebagai satu kesatuan data.                    |

**Kesimpulan:**  
Mengikuti prinsip Gestalt dalam visualisasi data membantu audiens memahami informasi dengan cepat, mengurangi kebingungan, dan menonjolkan pesan utama dari data.

### Menghilangkan Elemen yang Rumit & Cognitive Load

#### Menghilangkan Elemen yang Rumit
Bayangkan sebuah halaman kosong, lalu tambahkan satu elemen visual dan minta orang lain menjelaskan maknanya. Tambahkan elemen baru, ulangi prosesnya. Semakin banyak elemen yang ditambahkan, semakin banyak pula bagian yang terlewat oleh audiens. Setiap elemen membutuhkan waktu dan usaha otak untuk dipahami. Karena kapasitas otak terbatas, semakin banyak elemen berarti semakin tidak efektif visualisasi data.  
**Tips:** Identifikasi dan hapus elemen yang kurang efektif agar visual tetap rapi dan mudah dipahami.

---

#### Cognitive Load (Beban Kognitif)
Beban kognitif adalah usaha mental yang diperlukan untuk memproses dan mempelajari informasi. Otak manusia memiliki kapasitas pemrosesan terbatas, sehingga visualisasi data harus dibuat sesederhana mungkin agar mudah dipahami. Jika audiens membutuhkan waktu lama untuk memahami data, berarti visualisasi perlu diperbaiki.  
**Faktor penting:** Usaha audiens dalam mencerna informasi dan tingkat kepercayaan terhadap data yang disajikan.

---

#### Kerumitan (Clutter)
Kerumitan atau clutter adalah elemen visual yang tidak menambah pemahaman. Clutter menyebabkan informasi sulit dipahami dan pengalaman membaca menjadi tidak nyaman. Visual yang rumit membuat audiens membuang waktu untuk memahami data, sehingga pesan utama bisa tidak tersampaikan dengan efektif.  
**Solusi:** Reduksi clutter untuk meningkatkan efektivitas komunikasi visual.


### Ringkasan: Preattentive Attributes untuk Fokus Visual

**Preattentive attributes** adalah elemen visual yang ditangkap oleh otak secara cepat dan tanpa sadar. Penggunaannya secara strategis membantu mengarahkan perhatian audiens ke bagian terpenting dari visualisasi data, sehingga pesan utama dapat tersampaikan secara efektif dan cepat.

Tujuan utama penggunaan atribut ini adalah:
1.  **Mengarahkan Fokus**: Membantu audiens untuk langsung melihat informasi kunci yang ingin disampaikan.
2.  **Menciptakan Hierarki Visual**: Mengatur elemen visual berdasarkan tingkat kepentingannya, memandu audiens dalam memproses informasi secara terstruktur.

Atribut ini dapat diterapkan pada teks (misalnya, menggunakan **bold**, *italic*, atau ukuran font berbeda untuk judul dan subjudul) maupun pada grafik untuk menonjolkan data tertentu.

#### Tabel Atribut Preatentif Utama

| Atribut | Penjelasan | Contoh Penerapan dalam Visualisasi |
| :--- | :--- | :--- |
| **Warna** | Gunakan warna secara selektif untuk menyoroti data atau informasi penting. Hindari penggunaan banyak warna yang tidak perlu. | Memberi warna cerah pada satu bar di bar chart untuk menarik perhatian, sementara bar lainnya berwarna netral (misal: abu-abu). |
| **Ukuran** | Ukuran relatif menunjukkan tingkat kepentingan. Elemen yang lebih besar akan dianggap lebih penting. | Membuat judul grafik lebih besar dari label. Dalam scatter plot, titik data yang paling signifikan dibuat lebih besar. |
| **Posisi** | Tempatkan informasi paling penting di posisi yang paling sering dilihat pertama kali oleh audiens (umumnya kiri atas). | Meletakkan kesimpulan atau insight utama di bagian atas atau pojok kiri atas visualisasi. |

Dengan memanfaatkan atribut-atribut ini, sebuah visualisasi data dapat berubah dari sekadar menyajikan data (eksploratif) menjadi menjelaskan sebuah cerita atau insight (eksplanatoris) secara efektif.

### Ringkasan: Konsep Desain dalam Komunikasi Data

Untuk membuat visualisasi yang efektif, penting untuk menerapkan prinsip-prinsip desain. Tiga konsep utama yang perlu diperhatikan adalah *affordances*, *accessibility*, dan *aesthetics*. Tujuannya adalah untuk menciptakan visual yang tidak hanya informatif tetapi juga intuitif, mudah dipahami, dan profesional.

| Konsep | Penjelasan Singkat | Tips Penerapan |
| :--- | :--- | :--- |
| **Affordances** | Menggunakan isyarat visual yang sudah umum dikenal agar audiens secara intuitif tahu cara membaca dan berinteraksi dengan grafik. | - **Warna:** Gunakan warna yang umum (misal: merah untuk negatif, hijau untuk positif).<br>- **Konsistensi:** Jaga konsistensi warna dan posisi elemen (judul, kesimpulan) di semua visualisasi. |
| **Accessibility** | Memastikan visualisasi dapat dipahami oleh audiens dari berbagai latar belakang dan tingkat keahlian, bukan hanya oleh para ahli. | - **Bahasa:** Gunakan bahasa yang sederhana dan hindari jargon.<br>- **Kejelasan:** Pastikan visual dapat menjelaskan dirinya sendiri. Tambahkan teks jika perlu untuk memperjelas pesan. |
| **Aesthetics** | Menciptakan visual yang menarik dan profesional melalui pilihan warna, bentuk, dan posisi yang harmonis untuk membangun kepercayaan audiens. | - **Kerapian:** Hindari elemen yang tidak perlu (clutter), seperti warna berlebihan atau garis yang mengganggu.<br>- **Harmoni:** Pastikan alignment rata dan palet warna selaras agar visual terlihat profesional. |

### Rangkuman dari Sub-modul Transformasi Data ke Visual

Diagram adalah representasi visual dari informasi yang digunakan untuk mempermudah pemahaman data.

---

#### Bentuk Grafik dalam Visualisasi Data
- **Bar/Column Chart**: Mewakili kategori dengan persegi panjang yang panjangnya sebanding dengan nilai data.
- **Line Chart**: Menampilkan data sebagai serangkaian titik ('penanda') yang dihubungkan oleh garis lurus.
- **Area Chart**: Mirip Line Chart, namun area di bawah garis diisi dengan warna.
- **Pie Chart**: Grafik lingkaran yang merepresentasikan 100% dari keseluruhan, dengan irisan sebagai bagiannya.
- **Scatter Chart**: Menggunakan titik untuk merepresentasikan nilai pada sumbu horizontal dan vertikal.
- **Map Chart**: Memvisualisasikan data pada peta geografis, di mana warna menunjukkan intensitas data.

---

#### Kesalahan Umum dalam Visualisasi Data
- **Nilai persentase tidak sesuai** (total tidak 100% pada pie chart).
- **Terlalu banyak data** yang ingin ditampilkan.
- **Tidak mengikuti standar penulisan** (misalnya, sumbu Y terbalik).
- **Sumbu yang terpotong** (tidak dimulai dari 0).
- **Penggunaan grafik 3D yang tidak sesuai** sehingga menimbulkan bias.
- **Grafik susah dibandingkan** karena informasi tidak jelas.

---

#### Data Provenance (Dokumentasi Data)
Dokumentasi data (provenance) adalah proses mencatat sumber dan riwayat data untuk memastikan keabsahan, validitas, dan kepercayaan.
- **Contoh**: Metadata, *version history*.
- **Pendekatan**: Dicatat dalam teks (README), disimpan dalam sistem, dan ditulis dalam format yang dapat dibaca mesin/manusia.
- **Bentuk Dokumentasi**: Berkas README, *Data Dictionaries*, *Data Paper*.
- **Tips**: Gunakan alur terstruktur dan berikan informasi yang jelas serta detail.

---

#### Exploratory vs. Explanatory Analysis
- **Exploratory Analysis**: Membimbing audiens untuk menjelajahi data bersama.
- **Explanatory Analysis**: Langsung menyampaikan poin utama dan implikasi dari data.

**Tiga Poin Penting dalam Explanatory Analysis:**
1.  **Kenali Audiens**: Sesuaikan komunikasi dengan kebutuhan dan latar belakang audiens.
2.  **Pilih Cara Komunikasi**: Tentukan tingkat detail berdasarkan media (presentasi langsung vs. dokumen).
3.  **Gunakan Data untuk Menegaskan Maksud**: Manfaatkan metode 5W (What, Who, When, Where, Why) untuk memperjelas cerita.

---

#### Visualisasi Data yang Efektif
- **Teks**: Efektif untuk menampilkan satu atau dua angka penting.
- **Tabel**: Cocok untuk audiens beragam dengan banyak unit pengukuran.
- **Heatmap**: Menggabungkan detail tabel dengan visualisasi warna untuk menyoroti data penting.
- **Grafik**:
    - **Scatterplot**: Menunjukkan hubungan antara dua variabel.
    - **Line Chart**: Ideal untuk data kontinu (perubahan dari waktu ke waktu).
    - **Slopegraph**: Fokus pada perubahan antara dua titik.
    - **Bar Chart**: Mudah dibaca untuk membandingkan kategori.
    - **Stacked Bar Chart**: Membandingkan total dan sub-komponen.
    - **Waterfall Chart**: Menunjukkan perubahan bertahap dari nilai awal ke akhir.
    - **Area Chart**: Baik untuk menampilkan perbedaan nilai yang sangat besar.

---

#### Prinsip Gestalt dalam Persepsi Visual
- **Proximity (Kedekatan)**: Objek yang berdekatan dianggap satu kelompok.
- **Similarity (Kesamaan)**: Objek dengan warna/bentuk/ukuran sama dianggap satu kelompok.
- **Enclosure (Pembeda)**: Objek dalam satu area/border dianggap satu kelompok.
- **Closure (Penutupan)**: Otak melengkapi bentuk yang tidak utuh.
- **Continuity (Kesinambungan)**: Otak membentuk pola berkesinambungan.
- **Connection (Koneksi)**: Objek yang terhubung secara fisik dianggap satu kelompok.

---

#### Cognitive Load & Clutter
- **Cognitive Load (Beban Kognitif)**: Usaha mental yang diperlukan untuk memproses informasi. Visualisasi harus sederhana untuk mengurangi beban ini.
- **Clutter (Kerumitan)**: Elemen visual yang tidak menambah pemahaman. Identifikasi dan hapus clutter agar visualisasi efektif.

---

#### Preattentive Attributes
Atribut visual yang ditangkap otak secara cepat dan tanpa sadar. Gunakan untuk:
- **Mengarahkan Fokus**: Menyoroti informasi kunci.
- **Menciptakan Hierarki Visual**: Mengatur elemen berdasarkan tingkat kepentingan.
- **Contoh**: Penggunaan **warna** yang menonjol, **ukuran** yang lebih besar, **font tebal**, dan **posisi** strategis.

---

#### Konsep Desain dalam Komunikasi Data
1.  **Affordances**: Gunakan isyarat visual yang umum dikenal (misal: warna konsisten, posisi judul yang sama) agar audiens intuitif memahami grafik.
2.  **Accessibility**: Pastikan visualisasi dapat dipahami oleh semua orang dengan menggunakan bahasa sederhana dan visual yang dapat menjelaskan dirinya sendiri.
3.  **Aesthetics**: Ciptakan visual yang menarik dan profesional melalui harmoni warna, bentuk, dan posisi untuk membangun kepercayaan audiens.
